Dictionary in Python
Python dictionaries (often misspelled as "dictanary") store data as key-value pairs, offering fast lookups by unique keys. They are mutable, unordered (though insertion order is preserved since Python 3.7), and use curly braces {} for creation.

```python
#example
student ={
    "name":"John",
    "age":20,
    "grade":"A"
}

# print(student["name"])
print(student)
# print(student.keys())
# print(student.values())
# print(student.items())
print(type(student))
```

```python
empty_dict = dict();
print(empty_dict)

person = dict(name="Bob", age=30)
print(person)

pairs = [("name", "Charlie"), ("age", 35), ("city","Paris")]
print(pairs) #list

person2 = dict(pairs) #converts into dict
print(person2)

```

```python
#gives default value
keys = ["name", "age", "city"]
person3 = dict.fromkeys(keys, "Unknown")
print(person3)

default_dict2 = dict.fromkeys(keys)
print(default_dict2)
```

```python
#example
student ={
    "name":"Soumyajit Banerjee",
    "age":21,
    "CGPA":"8.9",
    "Stream":"Information Technology"
}
print(student)

print(student.keys())
print(student.values())
print(student.items())
print(student.get("name"))
print(student.get("grade")) #no error gives None
print(student.get("grade","Not Available")) #overrites the default None to Not Availabe

print()

#all are works with key
print("name" in student)
print("grade" in student)
print("grade" not in student)
print("Soumyait Banerjee" in student)

print(len(student))
```

```python
car = {
    "brand":"Toyata",
    "model":"Camry",
    "year":1964
}

#can be accessed like this way
print(car["brand"])
print(car["year"])

#add or update Key here with Color
car["Color"] = "Blue"

car["year"] = 2021 #can be update via this way

#update OR ADD a key value
car.update({"model":"Corolla","price":"$25000", "type":"SEADAN"})

print(car)

#delete key with its value
del car["Color"]

print(car)

# Returns value and deletes
print(f"Popped Value {car.pop("model")}")
print(car)
# Returns value and deletes with default Value error name bydefault its None
print(f"Price popped : {car.pop("price","Not Found")}")
print(car)
```

```python
dict1 = {"a":1, "b":2}
dict2 = {"c":3, "d":4}
dict1.update(dict2)
print(f"Using Update {dict1}")


dict3 = {"a":1, "b":2}
dict4 = {"c":3, "d":4,"b":1}
#handle duplicates whose value is comes last consider its
merged_dict = {**dict3,**dict4}
print(f"Using ** Operators : {merged_dict}")


dict4 = {"a":1, "b":2}
dict5 = {"c":3, "d":4}
merged_dict = dict4 | dict5
print(f"Using | Operators : {merged_dict}")

```

```python
student ={
    "name":"Soumyajit Banerjee",
    "age":21,
    "CGPA":"8.9",
    "Stream":"Information Technology"
}
for key in student:
    print(key)

print()

for key in student.keys():
    print(key)

print()

for value in student.values():
    print(value)

print()


for key,value in student.items():
  print(f"{key} : {value}")
```

```python
student = {
    "student1": {
        "name": "Soumyajit Banerjee",
        "age":21,
        "course":{
            "name":"B.Tech",
            "branch":"Information Technology"
        }
    },
    "student2": {
        "name": "Hamza Majari",
        "age":22,
        "course":{
            "name":"B.Tech",
            "branch":"Information Technology"
        }
    },
    "student3": {
        "name": "Saddam Hossian",
        "age":24,
        "course":{
            "name":"B.Tech",
            "branch":"Information Technology"
        }
    }
}
print(student)

#using get
print(student["student1"].get("name"))

#using[]
print("Student Name : ",student["student1"]["name"])
print("Student Age : ",student["student2"].get("age"))

print()

#for nested
for key, value in student.items():
    for key, value in value.items():
        print(f"{key} : {value}")
    print()
```

```python
numbers = [1,2,3,4,5]
squares = {num: num**2 for num in numbers}
cube = {num: num**3 for num in range(1,6)}
print(squares)
print(cube)

words = {"apple","banana","cherry"}
word_lengths = {word: len(word) for word in words}
print(word_lengths)
```

```python
keys = ["name", "age", "city"]
values = ["Bob", 30, "New York"]
#convert lists to dict
dicts = dict(zip(keys,values))
print(dicts)

numbers = range(1,11)
squares = {num: num**2 for num in numbers if num % 2 ==0}
print(squares)

words = ["hi","hello","hey","goodbye","yes"]
word_lengths = {word: len(word) for word in words if len(word)>4}
print(word_lengths)



student={
    "john":85,
    "emma":92,
    "alice":78,
    "bob":95
}
passed_students = {name: score  for name, score in student.items() if score >= 90}
print(passed_students)
```

```python
fruits = {"apple":"red","banana":"yellow","grape":"purple"}
upperCase = {key: value.upper() for key,value in fruits.items()}
print(upperCase)
```

```python
fruits = {"apple":1.00,"banana":0.50,"grape":0.75}

newPrices = {key : value*1.10 for key,value in fruits.items()}
print(newPrices)
```

```python
celsius = {"morning":20,"afternoon":25,"evening":30}
fh = {key: (value*9/5)+32 for key,value in celsius.items()}
print(fh)
```
