fruits.sort(key=str.lower): This is the core of the sorting. The sort() method sorts the list elements in ascending order by default. However, the key=str.lower argument modifies this behavior. It tells Python to convert each string to lowercase temporarily for comparison purposes only before sorting. The actual items in the list remain with their original casing. This ensures that 'Apple' and 'apple' are treated the same for sorting, leading to a case-insensitive sort.

```python
fruits = ["Apple","Banana","Cherry","Mango","elderberry","Banana"]
fruits.sort(key=str.lower)
print(fruits)
```

The map() function in Python applies a given function to each item in an iterable (like a list or tuple) and returns a map object, which is an iterator of the results. This makes it efficient for transforming data without explicit loops, promoting functional programming style.

```python
num = [1,2,3,4,5]

#map(func,iterable)
x = list(map(lambda x:x*2,num))
print(x)
```

The filter() function in Python constructs an iterator from elements of an iterable for which a function returns True. Like map(), it produces a lazy iterator, so convert it to a list or tuple to view results, making it memory-efficient for large datasets.

```python
num = [1,2,3,4,5]

def is_even(x):
  return x%2==0

x = list(filter(is_even,num))
print(x)

words = ['apple', 'bat', 'cat', 'ant']
y = list(filter(lambda w:w.startswith('a'),words))
print(y)

```

When filter(None, items) is used in Python, the None argument acts as a special shorthand that removes all falsy values from the iterable, keeping only truthy ones

```python
items = [0, 1, '', 'hello', [], [1, 2], None, True]
truthy = list(filter(None, items))
print(truthy)  # Output: [1, 'hello', [1, 2], True]
```

The reduce() function from Python's functools module applies a function cumulatively to items in an iterable, reducing them to a single value (unlike map() which transforms or filter() which selects). It performs a left-fold operation, often used for aggregations like sums or products.

```python
from functools import reduce

numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)  # 10

product = reduce(lambda x, y: x * y, numbers)
print(product)  # 24

#with initialize value here 100
numbers = [1, 2, 3]
result = reduce(lambda x, y: x + y, numbers, 100)
print(result)  # 106  (100+1+2+3)


#Find max/min:
numbers = [3, 1, 4, 1, 5]
max_val = reduce(lambda x, y: x if x > y else y, numbers)
print(max_val)  # 5

min_val = reduce(lambda x, y: x if x < y else y, numbers)
print(min_val)  # 1


```
